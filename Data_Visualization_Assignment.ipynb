 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a dataset
data = {'Category': ['Electronics', 'Clothing', 'Groceries', 'Furniture', 'Sports'],
        'Value': [250, 150, 300, 200, 180]}

df = pd.DataFrame(data)

# Visualization
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Value', data=df, palette='viridis')

# Adding titles and labels
plt.title('Distribution of Numerical Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```

 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Display plots in the notebook
output_notebook()

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label="X", y_axis_label="sin(X)", width=700, height=400)

# Here Adding a line representing the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# To Show the plot
show(p)
```
